### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
df.info()
df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
#missing_values = pd.DataFrame(missing_values[missing_values > 0].sort_values(ascending=False))
#missing_values.rename(columns={0:'count'},inplace=True)

missing_values
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(18,12))
sns.heatmap(df.isnull(), cmap='coolwarm', yticklabels=False)
plt.tight_layout()
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_employment = df['Employment'].isnull().sum()
missing_employment
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
freq_employment = df['Employment'].value_counts().idxmax()
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df['Employment'] = df['Employment'].fillna(freq_employment)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
plt.figure(figsize=(20,8))
employment_count = df['Employment'].value_counts()
employment_count.plot(kind='bar', color='skyblue')
plt.title('Distribution of Employment Status After Imputation')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
